# Document retrieval from Json data

```python
import turicreate
import pandas as pd
import datetime
import re
import operator
```

# Load some text data from Local Drive scrapped from Online


```python
df = pd.read_json('ArticleDataset.json')
df.head()
```

## Overview of Dataset

```python
df.describe()
```

 ## Take the query sentence from user and save it to the dataframe

```python
data = [{'date': datetime.date.today(),'title':'query','text':input(),'url':'N/A','4': 'N/A'}]
df.loc[len(df.index)]=list(data[0].values())
```

## Data Cleaning, Converting text column to lowecase
## Replacing Numbers from text Column
## Raplacing Html String 

```python
df = df.drop(columns=['4'])
df['text'] = df['text'].str.lower()
df['text'] = df['text'].str.replace('\d+', '')
df['text'] = df['text'].str.replace('{html}','')
df
```

# Explore data

```python
article = turicreate.SFrame(df)

article.show()
```

## Applying tfidf algorithm to the text colum and saving it to a new column

```python
article['tfidf'] = turicreate.text_analytics.tf_idf(article['text'])
```

```python
data = article.to_dataframe()
data
```

## Calculating distance using dataframe data Structure

```python
distance_dataframe = {}
find = len(data['tfidf']) -1

for i in range(len(data['tfidf'])):
    distance_dataframe[data['title'][i]] = (turicreate.distances.cosine(data['tfidf'][find], data['tfidf'][i]))
```

## Showing the top 100 article title related to the search query


```python
sorted_distance_dataframe = dict(sorted(distance_dataframe.items(), key=operator.itemgetter(1)))
n_items_dataframe = list(sorted_distance_dataframe.keys())
n_items_dataframe[1:101]
```

# Calculating cosine distance using Sframe data structure, taking more time

```python
distance = {}
find = len(article['tfidf']) -1

for i in range(len(article['tfidf'])):
    
   distance[article['title'][i]] = (turicreate.distances.cosine(article['tfidf'][find], article['tfidf'][i]))

```

# Showing the top 100 article titel that have nearest distance to the query 

```python
sorted_d = dict(sorted(distance.items(), key=operator.itemgetter(1)))
n_items = list(sorted_d.keys())
n_items[1:101]
```

```python
knn_model = turicreate.nearest_neighbors.create(article,features=['tfidf'],label='title')
```
